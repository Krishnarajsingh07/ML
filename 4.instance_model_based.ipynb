## __âœ… 1. Instance-Based Learning (Lazy Learning)__
Definition

Instance-based learning stores the training data and makes predictions by comparing new samples to stored examples.
It does not create an explicit model during training.

ğŸ‘‰ It is called lazy learning because it delays the generalization process until prediction time.

__ğŸ” How it works__

All training instances are stored.

When a new query comes:

Algorithm compares the query to stored instances.

Uses similarity/distance metrics (like Euclidean distance).

Predicts the output based on the most similar instances.

__ğŸ“Œ Key Features__

No explicit model built during training.

Training is fast (no learning).

Prediction is slow (needs to search through training data).

Performs local generalization (only for nearby instances).

__â­ Advantages__

Simple to understand and implement.

Adapts well to complex decision boundaries.

Works well with large training sets.

Good for non-linear problems.

__âŒ Disadvantages__

Slow prediction (high latency).

Requires lots of memory.

Sensitive to noise and irrelevant features.

Distance metrics may fail in high dimensions (curse of dimensionality).

ğŸ§ª Examples

k-Nearest Neighbors (kNN)

Locally Weighted Regression (LWR)

Kernel regression

Support Vector Machines with memory of support vectors (semi-instance-based)
