# üìä Exploratory Data Analysis (EDA) ‚Äì Brief Notes

## üîé Definition

Exploratory Data Analysis (EDA) is the process of analyzing and summarizing a dataset to understand its structure, patterns, and relationships before building a machine learning model.

---

## üéØ Purpose of EDA

- Understand dataset structure
- Identify missing values
- Detect outliers
- Find relationships between variables
- Select important features
- Prepare data for modeling

---

## üìå Types of EDA

### 1Ô∏è‚É£ Univariate Analysis
- Analyzing one variable at a time
- Used to understand distribution and spread

Examples:
- Histogram
- Boxplot
- Value counts
- Mean, Median, Mode

---

### 2Ô∏è‚É£ Bivariate Analysis
- Analyzing relationship between two variables

Examples:
- Scatter plot
- Bar chart
- Correlation analysis

---

### 3Ô∏è‚É£ Multivariate Analysis
- Analyzing more than two variables

Examples:
- Heatmap
- Pairplot
- Correlation matrix

---

# üìå Important Concepts

## Percentiles
- 25% values below Q1
- 50% values below Median
- 75% values below Q3

## IQR (Inter Quartile Range)
```
IQR = Q3 - Q1
```

## Outlier Detection Formula
```
Lower Limit = Q1 - 1.5 * IQR
Upper Limit = Q3 + 1.5 * IQR
```
