# <span style="color:red; font-weight:bold;">Multiple Linear Regression (MLR)</span>

---

## <span style="color:red; font-weight:bold;">1️⃣ Definition</span>
**Multiple Linear Regression (MLR)** is a technique to predict the **dependent variable (Y)**  
using **two or more independent variables (X₁, X₂, … Xₚ)**.

**General Equation:**
$
Y = b_0 + b_1 X_1 + b_2 X_2 + ... + b_p X_p + \epsilon
$

Where:  
- $Y$ → Dependent variable (target)  
- $X_1, X_2, ..., X_p$ → Independent variables (features)  
- $b_0$ → Intercept  
- $b_1, b_2, ..., b_p$ → Coefficients  
- $\epsilon$ → Error term

---

## <span style="color:red; font-weight:bold;">2️⃣ Difference: Simple vs Multiple Linear Regression</span>

| Feature | Simple Linear Regression | Multiple Linear Regression |
|---------|------------------------|---------------------------|
| Number of features | 1 | 2 or more |
| Equation | $Y = b_0 + b_1 X$ | $Y = b_0 + b_1 X_1 + b_2 X_2 + ... + b_p X_p$ |
| Complexity | Low | Higher |
| Use case | Single factor effect | Multiple factor effect |

---

## <span style="color:red; font-weight:bold;">3️⃣ Assumptions of MLR</span>
1. **Linearity:** Relationship between features and target is linear  
2. **Independence of errors**  
3. **Homoscedasticity:** Constant variance of errors  
4. **No multicollinearity:** Features should not be highly correlated  
5. **Normality of errors** (for inference)
