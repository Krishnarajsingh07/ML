### Converting Numerical Data into Categorical/Binary Form

Two important techniques are:

---

## 1. Discretization
Discretization is the process of converting continuous numerical values into discrete categories.  
It divides a continuous range into meaningful intervals or groups.

**Example:**
- Temperature values → "Low", "Medium", "High"
- Age → "Child", "Adult", "Senior"
- Marks -> Grade

**Use Cases:**
- Simplifying continuous data
- Preparing data for models that work better with categories
- To handle outliers
- To improve the value


# Types of Discretization

Discretization is the process of converting continuous numerical data into discrete categories.  
There are three main types:

---

## 1. Supervised Discretization
This method uses the **target/output variable** to decide how intervals (bins) should be created.  
It tries to form bins that best separate different classes.

### Examples:
- Decision Tree–based discretization
- Entropy-based discretization
- Chi-Square / Chi-Merge discretization

### When to Use:
- When you want bins that improve classification accuracy.
- When the relationship with the target variable matters.

---

## 2. Unsupervised Discretization
This does **not** use the target variable.  
It depends only on the **distribution** of the feature.

### Types:

### a) Equal-Width Binning
Divides the range of values into **equal-sized intervals**.  
Example: Age 0–100 → 4 bins of size 25  
- 0–25  
- 26–50  
- 51–75  
- 76–100

### b) Equal-Frequency Binning
Each bin contains the **same number of data points**.  
Example: 1000 records → 4 bins → 250 values per bin.

### c) Clustering-based Discretization
Uses algorithms like **k-means** to form groups based on similarity.

---

## 3. Custom / Domain-Driven Discretization
Bins are created using **real-world rules or expert knowledge**.

### Examples:
- Age groups: Child (0–12), Teen (13–19), Adult (20–59), Senior (60+)
- BMI categories: Underweight, Normal, Overweight, Obese
- Temperature: Cold, Warm, Hot

### When to Use:
- When interpretability is important.
- When domain knowledge gives meaningful categories.

---

# Summary Table

| Type                          | Uses Target? | Examples                                   | Best For                         |
|------------------------------|--------------|---------------------------------------------|----------------------------------|
| **Supervised Discretization** | ✔ Yes        | Decision tree, Chi-Merge                    | Improving model accuracy         |
| **Unsupervised Discretization** | ✘ No         | Equal-width, Equal-frequency, k-means       | General data preprocessing       |
| **Custom Discretization**     | ✘ No         | Age groups, BMI ranges                      | Human interpretability           |



---

## 2. Binarization
Binarization converts numerical or categorical values into **binary (0/1)** values.  
It creates a threshold and assigns:
- **1** if the value meets the condition  
- **0** otherwise  

**Example:**
- Convert age into a binary feature:  
  - Age ≥ 18 → 1 (Adult)  
  - Age < 18 → 0 (Minor)

# Equal-Width Binning (Uniform Binning)

## Definition
Equal-Width Binning — also called **Uniform Binning** — is a type of unsupervised discretization where the entire range of a numerical variable is divided into **intervals (bins) of equal size**.

This means:
- All bins have the same width.
- Only the *count* of data points inside each bin may vary.
- No change in speading of data

---

## How It Works
1. Find the minimum value (min)
2. Find the maximum value (max)
3. Decide the number of bins (k)
4. Compute bin width:

$$
\text{Bin Width} = \frac{\text{max} - \text{min}}{k}
$$

5. Create bins using this width.

---

## Example
Suppose Age ranges from 0 to 100, and we create 4 bins:

$$
\text{Bin Width} = \frac{\text{Max Value} - \text{Min Value}}{k}
$$

Bins will be:
- 0–25  
- 26–50  
- 51–75  
- 76–100  

Each bin has equal width (25).

---

## Advantages
- Very simple to apply  
- Easy to understand  
- Good for data with uniform distribution

---

## Disadvantages
- Highly sensitive to **outliers**  
- Bins may have **very uneven number of data points**  
- Not good for skewed distributions

---

## Summary
**Equal-Width Binning = Uniform Binning**  
Both terms mean the same thing → splitting data into intervals of equal size.



# Formula for Equal-Width (Uniform) Binning

To divide a continuous variable into **k equal-width bins**, use:

## 1. Bin Width Formula
$$
\text{Bin Width} = \frac{\text{Max Value} - \text{Min Value}}{k}
$$

---

## 2. Bin Boundaries
Once width is known, the bin boundaries are:

$$
\text{Bin}_i = 
\left[
\text{Min} + (i-1) \times \text{Width} \ ,\ 
\text{Min} + i \times \text{Width}
\right]
$$

where  
$$
i = 1, 2, 3, \dots, k
$$

---

## Example
If Min = 0, Max = 100, k = 4:

$$
\text{Width} = \frac{100 - 0}{4} = 25
$$

Bins:
- Bin₁: [0, 25]  
- Bin₂: [25, 50]  
- Bin₃: [50, 75]  
- Bin₄: [75, 100]

# Equal-Frequency Binning (Quantile Binning)

Equal-Frequency Binning is a type of unsupervised discretization where **each bin contains the same number of data points**.

This is also called:
- **Quantile Binning**
- **Percentile Binning**

---

## How It Works
1. Sort the data in ascending order.
2. Decide the number of bins (k).
3. Divide the sorted data into **k groups**, each containing approximately \( \frac{N}{k} \) data points.

---

## Formula

### 1. Number of points per bin
If the dataset has **N** values and we want **k bins**, then:

$$
\text{Points per bin} = \frac{N}{k}
$$

---

### 2. Bin boundaries (quantiles)
Bin boundaries are computed using quantiles:

$$
\text{Boundary}_i = Q\left(\frac{i}{k}\right)
$$

where  
- \( Q(p) \) = p-th quantile  
- \( i = 0, 1, 2, \dots, k \)

---

## Example

Dataset (sorted):  
4, 7, 10, 12, 15, 18, 20, 22  
Total values (N) = 8  
Bins (k) = 4  

Each bin should have:

$$
\text{Points per bin} = \frac{8}{4} = 2
$$

So the bins become:
- **Bin 1:** 4, 7  
- **Bin 2:** 10, 12  
- **Bin 3:** 15, 18  
- **Bin 4:** 20, 22  

Each bin has equal frequency = 2 values.

---

## Advantages
- Works very well for **skewed data**
- All bins get roughly similar number of samples
- Prevents empty bins

## Disadvantages
- Bin ranges may be uneven
- Harder to interpret compared to equal-width bins

# K-Means Binning (Clustering-Based Discretization)

K-Means Binning is a type of **unsupervised discretization** where bins are created using the
**k-means clustering algorithm** instead of fixed intervals.

Instead of slicing the data by equal width or equal frequency, k-means groups similar values
together based on distance.

---

## How It Works

1. Choose the number of clusters (k), which becomes the number of bins.
2. Apply the k-means algorithm to the numerical feature.
3. K-means finds cluster centers (centroids).
4. Each value is assigned to the nearest centroid.
5. These cluster groups become the **bins**.

---

## Example

Suppose we have values:

`[2, 3, 4, 10, 11, 12]`

If we choose k = 2, k-means may create clusters like:

- Cluster 1 (low values): 2, 3, 4  
- Cluster 2 (high values): 10, 11, 12  

Bin labels:
- Bin 0 → values near centroid 3  
- Bin 1 → values near centroid 11  

---

## Why Use K-Means for Binning?

- Works better when data is **not uniformly distributed**
- Automatically finds meaningful clusters
- Captures natural groupings in data
- Reduces noise by grouping similar values

---

## Advantages

- Adapts to the shape of the data  
- More flexible than equal-width or equal-frequency  
- Good for skewed or multi-modal distributions  

---

## Disadvantages

- Requires choosing k  
- Cluster boundaries are not interpretable like domain bins  
- Sensitive to outliers  
- Different runs may produce slightly different bins  

---

# Custom / Domain-Driven Discretization

Custom (or Domain-Driven) Discretization is a method where bins or categories are created using **expert knowledge, real-world logic, or domain-specific rules**, rather than mathematical formulas.

This method relies on understanding the **meaning** of the data, not just its numerical properties.

---

## How It Works
Instead of automatically computing bin boundaries, you manually decide intervals based on:
- Industry standards
- Medical rules
- Business logic
- Real-world thresholds
- Government guidelines
- Expert recommendations

---

## Examples

### 1. Age Groups
These bins are not equal-width or equal-frequency; they are meaningful categories:
- 0–12 → Child  
- 13–19 → Teen  
- 20–59 → Adult  
- 60+ → Senior

### 2. BMI Categories (Medical Standard)
- < 18.5 → Underweight  
- 18.5–24.9 → Normal  
- 25–29.9 → Overweight  
- ≥ 30 → Obese  

### 3. Blood Pressure Categories
- < 120/80 → Normal  
- 120–139 / 80–89 → Prehypertension  
- ≥ 140/90 → Hypertension  

### 4. Credit Score Bins
- < 580 → Poor  
- 580–669 → Fair  
- 670–739 → Good  
- 740–799 → Very Good  
- 800+ → Excellent  

These categories are **defined by domain rules**, not math.

---

## When to Use It
- When interpretability is important  
- When domain experts define meaningful ranges  
- When mathematical binning does not match real-world meaning  
- When the feature has known standard categories  

---

## Advantages
- Highly interpretable  
- Matches real-world logic  
- Works well for medical, financial, and business data  
- Makes results easier for humans to understand  

---

## Disadvantages
- Requires domain knowledge  
- Not always optimal for ML models  
- May not adapt well to unusual datasets  

---

## Summary
Custom / Domain-Driven Discretization =  
**Bins created using human logic + domain expertise**, not formulas.
