## __ðŸ“Œ What is a Tensor?__
__Definition:__ A tensor is a generalized multi-dimensional array that stores numerical data. It's the standard way to represent data in neural networks and deep learning frameworks.

__Analogy:__

+ Scalar = Single number (0D tensor)

+ Vector = List of numbers (1D tensor)

+ Matrix = Table of numbers (2D tensor)

+ Tensor = All of the above, extended to N dimensions

## __ðŸ”„ Tensor Dimensions with NumPy Examples__

__zero (0) D Tensor (scaler)__

```python
import numpy as np

a = np.array(4)
print(f"this is the value of this array {a}")
print(a)

print(f"The Dimension of this array is {np.ndim(a)}")
print(f"The shape of this array is {np.shape(a)}")
print(f"The rank of this array is {a.ndim}")
```
