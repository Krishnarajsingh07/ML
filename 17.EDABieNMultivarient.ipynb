# ðŸ“Œ Types of Variable Combinations

When analyzing two columns, there are 3 possible combinations:

# 1ï¸âƒ£ Numerical vs Numerical

### âœ… Best Plot: Scatter Plot

Used to check relationship between two numeric variables.

Example (Tips Dataset):
- Total Bill vs Tip

```python
sns.scatterplot(x="total_bill", y="tip", data=tips)
```

### ðŸ“Œ Insight:
- Shows linear/non-linear relationship
- Each dot represents one observation
- Helps detect trends & correlation

---

## ðŸ”¥ Multivariate Extension (Adding More Variables)

We can add extra dimensions using:

- `hue` â†’ for category (color)
- `style` â†’ marker style
- `size` â†’ size of marker

Example:

```python
sns.scatterplot(
    x="total_bill",
    y="tip",
    hue="sex",
    style="smoker",
    size="size",
    data=tips
)
```

Now the graph shows:
- Bill vs Tip
- Gender
- Smoker status
- Group size

ðŸ‘‰ This is Multivariate Analysis.

---

# 2ï¸âƒ£ Numerical vs Categorical

### âœ… Best Plot: Bar Plot

Used to compare average of numerical column across categories.

Example (Titanic):
- Pclass vs Age

```python
sns.barplot(x="Pclass", y="Age", data=titanic)
```

### ðŸ“Œ Insight:
- Shows mean value
- Helps compare groups
- Includes confidence interval (black line)

---

## Multivariate Version

Add `hue`:

```python
sns.barplot(x="Pclass", y="Age", hue="Sex", data=titanic)
```

Now comparing:
- Class
- Age
- Gender

---

# 3ï¸âƒ£ Categorical vs Categorical

### âœ… Method 1: Crosstab

```python
pd.crosstab(titanic["Pclass"], titanic["Survived"])
```

Shows frequency table.

---

### âœ… Heatmap Visualization

```python
sns.heatmap(pd.crosstab(titanic["Pclass"], titanic["Survived"]), annot=True)
```

âœ” Darker color = Higher value  
âœ” Helps identify patterns easily  

---

## ðŸ“Š Percentage Analysis (Better Approach)

Instead of raw counts, calculate percentage:

```python
titanic.groupby("Pclass")["Survived"].mean() * 100
```

âœ” Gives survival percentage  
âœ” More meaningful than raw counts  

---

# ðŸ“¦ Boxplot (Numerical vs Categorical)

Used to compare distribution across categories.

```python
sns.boxplot(x="Sex", y="Age", data=titanic)
```

Shows:
- Min
- Q1 (25%)
- Median
- Q3 (75%)
- Max
- Outliers

---

# ðŸ“ˆ Distribution Plot (Comparing Groups)

Used to compare distributions.

Example:
Age distribution of Survived vs Not Survived

```python
sns.kdeplot(titanic[titanic["Survived"]==0]["Age"])
sns.kdeplot(titanic[titanic["Survived"]==1]["Age"])
```

âœ” Shows probability density  
âœ” Helps detect age-based survival patterns  

---

# ðŸ“Š Pairplot (Multiple Numerical Columns)

If dataset has multiple numeric columns:

```python
sns.pairplot(iris)
```

âœ” Automatically creates scatter plots for all combinations  
âœ” Diagonal shows histogram  

Add hue:

```python
sns.pairplot(iris, hue="species")
```

âœ” Helps visualize class separation  

---

# ðŸ“ˆ Line Plot (Time-Based Data)

Used when X-axis is time.

Example (Flights Dataset):

```python
sns.lineplot(x="year", y="passengers", data=flights)
```

âœ” Shows trend over time  
âœ” Best for time-series analysis  

---

# ðŸ“Š Pivot Table & Heatmap (Time Data)

```python
pt = flights.pivot_table(
    values="passengers",
    index="month",
    columns="year"
)

sns.heatmap(pt)
```

âœ” Shows monthly/yearly trends  
âœ” Darker color = More passengers
