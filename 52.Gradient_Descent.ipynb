# üìâ Gradient Descent in Machine Learning

## ‚úÖ What is Gradient Descent?

Gradient Descent is an optimization algorithm used to find the minimum value of a loss (error) function.

üëâ In Machine Learning, it is used to find best parameters (weights) that minimize error.

In simple words:  
It moves step-by-step in the direction where error decreases fastest.

---

## ‚úÖ Cost Function (Linear Regression)

$J(\theta) = \frac{1}{2m}\sum_{i=1}^{m}(h_\theta(x_i) - y_i)^2$

Where:  
- $m$ = number of data points  
- $h_\theta(x)$ = predicted output  
- $y$ = actual output  

---

## ‚úÖ Gradient Descent Update Rule

$\theta = \theta - \alpha \frac{\partial J(\theta)}{\partial \theta}$

Where:  
- $\theta$ = model parameter (weight)  
- $\alpha$ = learning rate  
- Gradient = slope of loss function  

---

## ‚úÖ Learning Rate ($\alpha$)

Learning rate controls step size.

| Learning Rate | Effect |
|--------------|--------|
| Too small | Slow training |
| Too large | May not converge |
| Good value | Fast convergence |

---

## ‚úÖ Five Types of Gradient Descent

### 1. Batch Gradient Descent
- Uses entire dataset for each update  
- Stable and accurate  
- Very slow for large data  

---

### 2. Stochastic Gradient Descent (SGD)
- Uses only **one data point** per update  
- Very fast  
- Loss fluctuates a lot  

---

### 3. Mini-Batch Gradient Descent
- Uses small batch (e.g., 32 or 64 samples)  
- Fast and stable  
- Most commonly used in practice  

---

### 4. Momentum Gradient Descent
- Adds previous gradient to current update  
- Speeds up learning  
- Reduces oscillations  

Update rule:  
$v = \beta v + \alpha \nabla J(\theta)$  
$\theta = \theta - v$

---

### 5. Adaptive Gradient Descent (Adam / Adagrad)
- Learning rate changes automatically  
- Works well for sparse data  
- Used in deep learning  

Adam update idea:  
Uses moving average of gradients and squared gradients.

---

## ‚úÖ Algorithm Steps

1. Initialize weights randomly  
2. Predict output  
3. Calculate loss  
4. Compute gradient  
5. Update weights using $\theta = \theta - \alpha \nabla J(\theta)$  
6. Repeat until minimum loss

---

## ‚úÖ Advantages

- Easy to implement  
- Works with large datasets  
- Backbone of deep learning training  

---

## ‚ùå Disadvantages

- Sensitive to learning rate  
- May get stuck in local minima  
- Needs many iterations  

---

## ‚úÖ Very Short Exam Answer

Gradient Descent is an optimization algorithm to minimize loss function.  
It updates parameters in direction of negative gradient.  
Learning rate $\alpha$ controls step size.  
Five types are Batch, SGD, Mini-batch, Momentum and Adam.  
Used in regression and neural networks.

---

````python
class GDRegressor:

    def __init__(self,learning_rate,epochs):
        self.m = m
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,x,y):
        for i in range(self.epochs):
            loss_slope = -2 * np.sum( y - self.m*x.ravel() - self.b) 
            self.b = self.b - (self.lr * loss_slope)
            print(loss_slope,self.b)
        print('this is ' ,self.b)
````

````python
from sklearn.datasets import make_regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import cross_val_score

x , y = make_regression(n_samples=100,n_features=1,n_informative=1,n_targets=1,noise=20,random_state=13)
````

````python
plt.scatter(x,y)
````

````python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
````

````python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train,y_train)
print(lr.coef_)
print(lr.intercept_)
np.mean(cross_val_score(lr,x,y,scoring='r2',cv=10))
````

````python
y_pred = lr.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
````

````python
class GDRegressor:

    def __init__(self,learning_rate,epochs):
        self.m = 100
        self.b = -120
        self.lr = learning_rate
        self.epochs = epochs

    def fit(self,x,y):
        for i in range(self.epochs):
            loss_slope_b = -2 * np.sum( y - self.m*x.ravel() - self.b) 
            loss_slope_m = -2 * np.sum( (y - self.m*x.ravel() - self.b)*x.ravel()) 
            self.b = self.b - (self.lr * loss_slope_b)
            self.m = self.m - (self.lr * loss_slope_m)
        print('this is ' ,self.b)

    def predict(self,x):
        return self.m*x + self.b
````

````python
gd = GDRegressor(0.001,50)
gd.fit(x_train,y_train)
````

````python
gd.predict(x_train)
````

````python
y_pred = gd.predict(x_test)
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
````
