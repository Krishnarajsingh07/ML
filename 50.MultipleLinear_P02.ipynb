# **Multiple Linear Regression (MLR) â€“ Matrix Mathematical Formulation**

---

## **1ï¸âƒ£ Linear Model in Matrix Form**

For $n$ observations and $p$ independent variables:

$\mathbf{y} = \mathbf{X}\boldsymbol{\beta} + \boldsymbol{\varepsilon}$

---

## **2ï¸âƒ£ Definition of Matrices**

### **Design Matrix ($\mathbf{X}$)**

$\mathbf{X} \in \mathbb{R}^{n \times (p+1)}$

$\mathbf{X} =
\begin{bmatrix}
1 & x_{11} & x_{12} & \cdots & x_{1p} \\
1 & x_{21} & x_{22} & \cdots & x_{2p} \\
\vdots & \vdots & \vdots & & \vdots \\
1 & x_{n1} & x_{n2} & \cdots & x_{np}
\end{bmatrix}$

---

### **Target Vector ($\mathbf{y}$)**

$\mathbf{y} \in \mathbb{R}^{n \times 1}$

$\mathbf{y} =
\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n
\end{bmatrix}$

---

### **Parameter Vector ($\boldsymbol{\beta}$)**

$\boldsymbol{\beta} \in \mathbb{R}^{(p+1) \times 1}$

$\boldsymbol{\beta} =
\begin{bmatrix}
\beta_0 \\
\beta_1 \\
\vdots \\
\beta_p
\end{bmatrix}$

---

### **Error Vector ($\boldsymbol{\varepsilon}$)**

$\boldsymbol{\varepsilon} \in \mathbb{R}^{n \times 1}$

$\boldsymbol{\varepsilon} =
\begin{bmatrix}
\varepsilon_1 \\
\varepsilon_2 \\
\vdots \\
\varepsilon_n
\end{bmatrix}$

---

## **3ï¸âƒ£ Assumptions (Classical Linear Model)**

$E[\boldsymbol{\varepsilon}] = \mathbf{0}$  

$\text{Var}(\boldsymbol{\varepsilon}) = \sigma^2 \mathbf{I}$  

---

## **4ï¸âƒ£ Least Squares Objective Function**

Minimize the squared error:

$J(\boldsymbol{\beta}) = ||\mathbf{y} - \mathbf{X}\boldsymbol{\beta}||^2$

Expanded form:

$J(\boldsymbol{\beta}) = (\mathbf{y} - \mathbf{X}\boldsymbol{\beta})^T(\mathbf{y} - \mathbf{X}\boldsymbol{\beta})$

---

## **5ï¸âƒ£ Normal Equation (Matrix Solution)**

Taking derivative and setting to zero:

$\dfrac{\partial J}{\partial \boldsymbol{\beta}} = -2\mathbf{X}^T(\mathbf{y} - \mathbf{X}\boldsymbol{\beta}) = 0$

Solving:

$\mathbf{X}^T\mathbf{X}\boldsymbol{\beta} = \mathbf{X}^T\mathbf{y}$

---

## **6ï¸âƒ£ Closed-Form Estimator**

$\boxed{\boldsymbol{\beta} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}}$

(Provided $\mathbf{X}^T\mathbf{X}$ is invertible)

---

## **7ï¸âƒ£ Prediction in Matrix Form**

$\hat{\mathbf{y}} = \mathbf{X}\boldsymbol{\beta}$

For a new sample $\mathbf{x}_{new}$:

$\hat{y} = \mathbf{x}_{new}^T \boldsymbol{\beta}$

---

## **8ï¸âƒ£ Residual Vector**

$\mathbf{e} = \mathbf{y} - \hat{\mathbf{y}}$

---

## **9ï¸âƒ£ Variance of Coefficient Estimates**

$\text{Var}(\boldsymbol{\beta}) = \sigma^2(\mathbf{X}^T\mathbf{X})^{-1}$

---

## **ðŸ”Ÿ Compact Mathematical Statement**

> Multiple Linear Regression solves the optimization problem  
> $\min_{\boldsymbol{\beta}} ||\mathbf{y} - \mathbf{X}\boldsymbol{\beta}||^2$  
> with solution $\boldsymbol{\beta} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}$.
