# ðŸ“Œ Mini-Batch Gradient Descent (Mini GD)

---

## ðŸ”¹ Definition

**Mini-Batch Gradient Descent** is an optimization algorithm that updates model parameters using a **small subset (mini-batch)** of the training data instead of the entire dataset or a single data point.

---

## ðŸ”¹ Update Rule

For a mini-batch \( B \) of size \( b \):

$$
\theta = \theta - \alpha \nabla J_B(\theta)
$$

where:
- $( \theta $) = model parameters  
- $( \alpha $) = learning rate  
- $( \nabla J_B(\theta) $) = gradient computed on mini-batch $( B $)

---

## ðŸ”¹ How It Works

1. Shuffle the dataset  
2. Divide data into mini-batches of size \( b \)  
3. For each mini-batch:
   - Compute predictions  
   - Compute loss  
   - Compute gradients  
   - Update parameters  

---

## ðŸ”¹ Common Mini-Batch Sizes

$$
b \in \{16, 32, 64, 128, 256\}
$$

---

## ðŸ”¹ Why Mini-Batch GD is Used

- Faster than Batch GD  
- More stable than SGD  
- Efficient use of **CPU/GPU parallelism**  

---

## ðŸ”¹ Advantages

- Balanced trade-off between **speed** and **stability**
- Less noisy updates compared to SGD
- Scales well to large datasets
- Default choice in deep learning

---

## ðŸ”¹ Disadvantages

- Requires tuning of batch size
- Slightly noisier than Batch GD
- More complex than pure SGD

---

## ðŸ”¹ Comparison Snapshot

| Method | Data per Update | Stability | Speed |
|------|----------------|-----------|-------|
| Batch GD | All data | High | Slow |
| SGD | 1 sample | Low | Fast |
| **Mini GD** | Small batch | Medium | Fast |

---

## ðŸ§  One-Line Exam Answer

> **Mini-Batch Gradient Descent updates parameters using a small group of training samples, combining the stability of Batch GD and the speed of SGD.**

---

## ðŸ“Œ Practical Note

> **Almost all modern ML and DL frameworks (TensorFlow, PyTorch) use Mini-Batch Gradient Descent by default.**


# ðŸ“Š Difference Between Batch GD, Stochastic GD, and Mini-Batch GD

---

## 1ï¸âƒ£ Batch Gradient Descent (BGD)

### ðŸ”¹ Definition
- Computes gradient using **all training samples** before updating parameters.

### ðŸ”¹ Update Rule
$$
\theta = \theta - \alpha \nabla J(\theta)
$$

### ðŸ”¹ Characteristics
- One update **per epoch**
- Smooth and stable convergence
- Computationally expensive for large datasets

---

## 2ï¸âƒ£ Stochastic Gradient Descent (SGD)

### ðŸ”¹ Definition
- Updates parameters using **one randomly selected sample** at a time.

### ðŸ”¹ Update Rule
$$
\theta = \theta - \alpha \nabla J_i(\theta)
$$

### ðŸ”¹ Characteristics
- Many updates per epoch
- Fast but noisy learning
- Suitable for very large datasets

---

## 3ï¸âƒ£ Mini-Batch Gradient Descent (MBGD)

### ðŸ”¹ Definition
- Updates parameters using a **small batch** of samples.

### ðŸ”¹ Update Rule
$$
\theta = \theta - \alpha \nabla J_B(\theta)
$$

where  
$B$ = mini-batch size (e.g., 32, 64, 128)

---

## ðŸ“‹ Comparison Table

| Feature | Batch GD | Stochastic GD | Mini-Batch GD |
|-------|----------|---------------|---------------|
| Data per update | All samples | 1 sample | Small batch |
| Batch size | $n$ | $1$ | $b$ (32â€“256) |
| Updates per epoch | $1$ | $n$ | $n/b$ |
| Speed | Slow | Very fast | Fast |
| Memory usage | High | Very low | Medium |
| Convergence | Smooth | Noisy | Balanced |
| Scalability | Poor | Excellent | Excellent |
| Hardware friendly | âŒ | âŒ | âœ… (GPU) |

---

## âœ… When to Use

| Method | Best Used When |
|------|---------------|
| Batch GD | Small dataset, exact convergence |
| Stochastic GD | Very large or streaming data |
| Mini-Batch GD | Most practical ML & DL tasks |

---

## ðŸ§  One-Line Exam Summary

> **Batch GD is accurate but slow, Stochastic GD is fast but noisy, and Mini-Batch GD provides the best balance of speed and stability.**

---

## ðŸ“Œ Industry Standard

> **Mini-Batch Gradient Descent is the default choice in modern machine learning and deep learning frameworks.**


## __Code From scratch__

```python
from sklearn.datasets import load_diabetes
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.linear_model import LinearRegression
```

```python
x,y = load_diabetes(return_X_y=True)
x_train,x_test,y_train,y_test =  train_test_split(x,y,train_size=0.8,random_state=2)
```

```python
reg = LinearRegression()
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
r2_score(y_test,y_pred)
```

```python
import random

class MBRegression:

    def __init__(self,batch_size,learning_rate = 0.01, epochs = 100):

        self.coef__ = None
        self.intercept__ = None
        self.lr = learning_rate
        self.epochs = epochs
        self.batch_size = batch_size

    def fit(self,x_train,y_train):
        self.intercept__ =  0
        self.coef__ = np.ones(x_train.shape[1])
        
        for i in range(self.epochs):
            for j in range(int(x_train.shape[0]/self.batch_size)):
                
                index = random.sample(range(x_train.shape[0]),self.batch_size)

                y_hat = np.dot(x_train[index],self.coef__) + self.intercept__

                intercept_der = -2 * np.mean(y_train[index] - y_hat)
                self.intercept__ = self.intercept__ - (self.lr * intercept_der)

                coef_der  = -2 * np.dot((y_train[index] - y_hat),x_train[index])
                self.coef__ = self.coef__ - (self.lr * coef_der)

        print(self.intercept__,self.coef__)

    def predcit(self,x_test):
        return np.dot(x_test,self.coef__) + self.intercept__
```
