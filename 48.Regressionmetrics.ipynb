# <span style="color:red; font-weight:bold;">1.üëçMean Absolute Error (MAE)</span>

## <span style="color:red; font-weight:bold;">Definition</span>
**Mean Absolute Error (MAE)** is a regression evaluation metric that measures the  
**average absolute difference** between actual values and predicted values.  
It represents how far predictions are from the true values on average.

---

## <span style="color:red; font-weight:bold;">Mathematical Formula</span>
$
MAE = \frac{1}{n}\sum_{i=1}^{n} |y_i - \hat{y}_i|
$

### <span style="color:red; font-weight:bold;">Where:</span>
- $n$ ‚Üí number of observations  
- $y_i$ ‚Üí actual value  
- $\hat{y}_i$ ‚Üí predicted value  

---

## <span style="color:red; font-weight:bold;">Key Characteristics</span>
- Uses **absolute errors** instead of squared errors  
- **Less sensitive to outliers** compared to MSE and RMSE  
- Measured in the **same unit as the target variable**  
- All errors are treated equally  

---

## <span style="color:red; font-weight:bold;">Interpretation</span>
- **Lower MAE ‚áí Better model performance**
- $MAE = 0$ indicates a **perfect prediction**
- Example:  
  - If $MAE = 4$, predictions are off by **4 units on average**

---

## <span style="color:red; font-weight:bold;">Advantages</span>
- Simple and intuitive to understand  
- Unit is same as the unit of gien data
- Robust against outliers  
- Easy real-world interpretation  

---

## <span style="color:red; font-weight:bold;">Disadvantages</span>
- Does not penalize large errors strongly  
- Less sensitive to extreme mistakes  
- __Not differentiable at zero__  

---

## <span style="color:red; font-weight:bold;">When to Use MAE</span>
- When outliers should not dominate evaluation  
- When equal importance is given to all errors  
- When interpretability matters  

---

## <span style="color:red; font-weight:bold;">Comparison with Other Metrics</span>

| Metric | Penalizes Large Errors | Unit |
|------|------------------------|------|
| MAE | ‚ùå No | Same as target |
| MSE | ‚úÖ Yes | Squared |
| RMSE | ‚úÖ Yes | Same as target |

---

## <span style="color:red; font-weight:bold;">Python Example</span>
```python
from sklearn.metrics import mean_absolute_error

y_true = [3, 5, 7, 9]
y_pred = [2, 5, 8, 10]

mae = mean_absolute_error(y_true, y_pred)
print("MAE:", mae)

# <span style="color:red; font-weight:bold;">2.üëçMean Squared Error (MSE)</span>

## <span style="color:red; font-weight:bold;">Definition</span>
**Mean Squared Error (MSE)** is a regression evaluation metric that measures the  
**average of the squared differences** between actual values and predicted values.  
It penalizes larger errors more heavily due to squaring.

---

## <span style="color:red; font-weight:bold;">Mathematical Formula</span>
$
MSE = \frac{1}{n}\sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$

### <span style="color:red; font-weight:bold;">Where:</span>
- $n$ ‚Üí number of observations  
- $y_i$ ‚Üí actual value  
- $\hat{y}_i$ ‚Üí predicted value  

---

## <span style="color:red; font-weight:bold;">Key Characteristics</span>
- Uses **squared errors**, amplifying large mistakes  
- **Highly sensitive to outliers**  
- Always **non-negative**  
- Unit is the **square of the target variable‚Äôs unit**

---

## <span style="color:red; font-weight:bold;">Interpretation</span>
- **Lower MSE ‚áí Better model performance**
- $MSE = 0$ indicates a **perfect model**
- Large errors have a **disproportionately large impact**

---

## <span style="color:red; font-weight:bold;">Advantages</span>
- Strongly penalizes large errors  
- Differentiable everywhere ‚Üí useful for optimization  
- Widely used in regression algorithms  

---

## <span style="color:red; font-weight:bold;">Disadvantages</span>
- Very sensitive to outliers  
- Difficult to interpret due to squared units  
- Not robust when data contains noise  

---

## <span style="color:red; font-weight:bold;">When to Use MSE</span>
- When large errors must be heavily penalized  
- When model training requires a smooth loss function  
- When outliers are meaningful and important  

---

## <span style="color:red; font-weight:bold;">Comparison with Other Metrics</span>

| Metric | Sensitivity to Outliers | Unit |
|------|------------------------|------|
| MAE | Low | Same as target |
| MSE | High | Squared |
| RMSE | High | Same as target |

---

## <span style="color:red; font-weight:bold;">Python Example</span>
```python
from sklearn.metrics import mean_squared_error

y_true = [3, 5, 7, 9]
y_pred = [2, 5, 8, 10]

mse = mean_squared_error(y_true, y_pred)
print("MSE:", mse)

# <span style="color:red; font-weight:bold;">Root Mean Squared Error (RMSE)</span>

## <span style="color:red; font-weight:bold;">Definition</span>
**Root Mean Squared Error (RMSE)** is a regression evaluation metric that represents  
the **square root of the Mean Squared Error (MSE)**.  
It measures the average magnitude of prediction errors in the **same unit as the target variable**.

---

## <span style="color:red; font-weight:bold;">Mathematical Formula</span>
$
RMSE = \sqrt{\frac{1}{n}\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
$

### <span style="color:red; font-weight:bold;">Where:</span>
- $n$ ‚Üí number of observations  
- $y_i$ ‚Üí actual value  
- $\hat{y}_i$ ‚Üí predicted value  

---

## <span style="color:red; font-weight:bold;">Key Characteristics</span>
- Derived from **MSE** by taking the square root  
- **Highly sensitive to outliers**  
- Error is expressed in the **same unit as the target**  
- Penalizes large errors more than MAE  

---

## <span style="color:red; font-weight:bold;">Interpretation</span>
- **Lower RMSE ‚áí Better model performance**
- $RMSE = 0$ indicates **perfect predictions**
- Example:  
  - $RMSE = 6$ means predictions deviate by **6 units on average**

---

## <span style="color:red; font-weight:bold;">Advantages</span>
- Easier to interpret than MSE  
- Strong penalty for large errors  
- Commonly used for model evaluation  

---

## <span style="color:red; font-weight:bold;">Disadvantages</span>
- Very sensitive to outliers  
- Large errors dominate the metric  
- Less robust than MAE  

---

## <span style="color:red; font-weight:bold;">When to Use RMSE</span>
- When large errors are undesirable  
- When interpretability in original units is needed  
- When data has minimal extreme outliers  

---

## <span style="color:red; font-weight:bold;">Comparison with Other Metrics</span>

| Metric | Sensitivity to Outliers | Unit |
|------|------------------------|------|
| MAE | Low | Same as target |
| RMSE | High | Same as target |
| MSE | High | Squared |

---

## <span style="color:red; font-weight:bold;">Python Example</span>
```python
from sklearn.metrics import mean_squared_error
import numpy as np

y_true = [3, 5, 7, 9]
y_pred = [2, 5, 8, 10]

rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print("RMSE:", rmse)

# <span style="color:red; font-weight:bold;">R¬≤ Score (Coefficient of Determination)</span>

## <span style="color:red; font-weight:bold;">Definition</span>
**R¬≤ Score** measures how well the regression model explains the **variance in the dependent variable**.  
It compares the model‚Äôs predictions with a simple baseline model that always predicts the mean.

---

## <span style="color:red; font-weight:bold;">Mathematical Formula</span>
$
R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}
$

### <span style="color:red; font-weight:bold;">Where:</span>
- $n$ ‚Üí number of observations  
- $y_i$ ‚Üí actual value  
- $\hat{y}_i$ ‚Üí predicted value  
- $\bar{y}$ ‚Üí mean of actual values  

---

## <span style="color:red; font-weight:bold;">Key Characteristics</span>
- Dimensionless (no units)  
- Measures **goodness of fit**  
- Compares model performance against mean prediction  
- Can be **negative** for poor models  

---

## <span style="color:red; font-weight:bold;">Interpretation</span>
- $R^2 = 1$ ‚Üí perfect fit  
- $R^2 = 0$ ‚Üí model predicts as well as the mean  
- $R^2 < 0$ ‚Üí model performs worse than mean  
- Example:  
  - $R^2 = 0.85$ ‚Üí model explains **85% of variance**

---

## <span style="color:red; font-weight:bold;">Advantages</span>
- Easy to interpret  
- Widely used for model comparison  
- Indicates explanatory power  

---

## <span style="color:red; font-weight:bold;">Disadvantages</span>
- Does not indicate prediction error magnitude  
- Increases automatically with more features  
- Not reliable for non-linear models  

---

## <span style="color:red; font-weight:bold;">When to Use R¬≤</span>
- When measuring variance explained is important  
- When comparing models with the same target variable  
- For overall model fit evaluation  

---

## <span style="color:red; font-weight:bold;">Comparison with Error Metrics</span>

| Metric | Measures Error | Measures Fit |
|------|--------------|-------------|
| MAE | ‚úÖ Yes | ‚ùå No |
| RMSE | ‚úÖ Yes | ‚ùå No |
| R¬≤ | ‚ùå No | ‚úÖ Yes |

---

## <span style="color:red; font-weight:bold;">Python Example</span>
```python
from sklearn.metrics import r2_score

y_true = [3, 5, 7, 9]
y_pred = [2, 5, 8, 10]

r2 = r2_score(y_true, y_pred)
print("R¬≤ Score:", r2)

# <span style="color:red; font-weight:bold;">Adjusted R¬≤ Score</span>

## <span style="color:red; font-weight:bold;">Definition</span>
**Adjusted R¬≤** is a modified version of the **R¬≤ Score** that accounts for the  
**number of independent variables (features)** in the model.  
It penalizes the inclusion of irrelevant features and prevents overfitting.

---

## <span style="color:red; font-weight:bold;">Mathematical Formula</span>
$
Adjusted\ R^2 = 1 - (1 - R^2)\frac{n - 1}{n - p - 1}
$

### <span style="color:red; font-weight:bold;">Where:</span>
- $n$ ‚Üí number of observations  
- $p$ ‚Üí number of independent variables (features)  
- $R^2$ ‚Üí coefficient of determination  

---

## <span style="color:red; font-weight:bold;">Key Characteristics</span>
- Adjusts R¬≤ based on model complexity  
- Can **increase or decrease** when a feature is added  
- Penalizes unnecessary predictors  
- Always $\leq R^2$

---

## <span style="color:red; font-weight:bold;">Interpretation</span>
- Higher Adjusted $R^2$ ‚áí better model with optimal features  
- If a new feature improves the model, Adjusted $R^2$ **increases**  
- If a feature is irrelevant, Adjusted $R^2$ **decreases**  

---

## <span style="color:red; font-weight:bold;">Advantages</span>
- Prevents misleading improvement in R¬≤  
- Useful for **multiple linear regression**  
- Helps in feature selection  

---

## <span style="color:red; font-weight:bold;">Disadvantages</span>
- Not meaningful for very small datasets  
- Less intuitive than R¬≤  
- Still does not measure prediction error  

---

## <span style="color:red; font-weight:bold;">When to Use Adjusted R¬≤</span>
- When comparing models with different numbers of features  
- In multiple linear regression problems  
- When avoiding overfitting is important  

---

## <span style="color:red; font-weight:bold;">Comparison with R¬≤</span>

| Metric | Penalizes Extra Features | Always Increases |
|------|-------------------------|----------------|
| R¬≤ | ‚ùå No | ‚úÖ Yes |
| Adjusted R¬≤ | ‚úÖ Yes | ‚ùå No |

---

## <span style="color:red; font-weight:bold;">Python Example</span>
```python
from sklearn.metrics import r2_score

n = 100   # number of samples
p = 5     # number of features

r2 = 0.82
adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("Adjusted R¬≤:", adjusted_r2)
