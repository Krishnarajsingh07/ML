## __9.Offline learning /Deployment__

+ In many ML workflows, models are trained offline (also called batch learning). This means the model learns from a fixed dataset and does not update itself in real time after deployment.

* Simple Example

+ Imagine a recommendation system (like Netflix or Amazon).
If trained offline, it won't learn from today's user behavior until the next retraining cycle.
So recommendations quickly become outdated.
