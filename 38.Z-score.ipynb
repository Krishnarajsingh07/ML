# üî¥ Outlier Removal Using Z-Score

---

## üìå What is Z-Score?
The **Z-score** tells how many **standard deviations** a data point is away from the **mean**.

---

## üìê Z-Score Formula
$$
Z = \frac{x - \mu}{\sigma}
$$

Where:
- $$ x = \text{data point} $$
- $$ \mu = \text{mean} $$
- $$ \sigma = \text{standard deviation} $$

---

## üö® Outlier Rule
$$
|Z| > 3 \Rightarrow \text{Outlier}
$$

---

## üß™ Steps to Remove Outliers (Z-Score Method)

### 1Ô∏è‚É£ Calculate Z-Scores  
### 2Ô∏è‚É£ Identify Outliers  
### 3Ô∏è‚É£ Remove Those Data Points  

---

## üß™ Python Implementation

```python
from scipy.stats import zscore

# Calculate Z-scores
df['z_score'] = zscore(df['Salary'])

# Remove outliers
df_clean = df[df['z_score'].abs() <= 3]

## üîî Bell Curve (Normal Distribution) ‚Äî Key Points

---

### üìå Shape & Symmetry
- Bell curve is **perfectly symmetric** about the mean (Œº).
- Left and right sides are **mirror images**.
- Most natural and real-world data follows this shape.

---

### üìå Mean, Median & Mode
- In a normal distribution:
  
  **Mean = Median = Mode**
  
- All lie at the **center peak** of the curve.

---

### üìå Center Partition (Very Important)
- Area between **Œº and +1œÉ** = **34.1%**
- Area between **‚àí1œÉ and Œº** = **34.1%**
- Total within **Œº ¬± 1œÉ** = **68.2%**

---

### üìå Empirical Rule (68‚Äì95‚Äì99.7 Rule)
- **68.2%** data lies within **Œº ¬± 1œÉ**
- **95.4%** data lies within **Œº ¬± 2œÉ**
- **99.7%** data lies within **Œº ¬± 3œÉ**

---

### üìå Area Under the Curve
- Total area under the bell curve = **1**
- Probability on each side of the mean = **0.5**

---

### üìå Z-Score Meaning
- Z-score measures how many standard deviations a value is from the mean.
- **Positive Z-score** ‚Üí value lies to the **right** of mean
- **Negative Z-score** ‚Üí value lies to the **left** of mean

---

### üìå Z-Score Formula
Z = (x ‚àí Œº) / œÉ

Where:
- x = data point  
- Œº = mean  
- œÉ = standard deviation  

---

### üìå Outlier Detection (Z-Score Method)
- In a normal distribution:
  
  **|Z| > 3 ‚Üí Outlier**
  
- Values beyond **¬±3œÉ** are extremely rare.

---

### üìå Tails of the Curve
- The bell curve **never touches the x-axis**.
- It approaches the x-axis **asymptotically**.

---

### üìå Standard Normal Distribution
- When:
  
  **Œº = 0**  
  **œÉ = 1**
  
- It is called the **standard normal distribution**.
- Used in **Z-tables** and probability calculations.

---

### üìå Effect of Parameters
- **Mean (Œº)** shifts the curve left or right.
- **Standard deviation (œÉ)** controls the spread (width).

---

### üìå Maximum Height
- The curve reaches its **maximum height at the mean (Œº)**.
- Probability density is highest at the center.

---

### üìå Applications of Bell Curve
- Exam scores
- IQ distribution
- Measurement errors
- Machine learning assumptions
- Statistical inference

---

### üéØ One-Line Exam Summary
A bell curve is a symmetric distribution where mean = median = mode, and 99.7% of data lies within three standard deviations of the mean.

````python
# Calculating z score

df['cgpa_z_score'] = (df['cgpa']-df['cgpa'].mean())/df['cgpa'].std()
````

````python
df[(df['cgpa_z_score']>3) | (df['cgpa_z_score']<-3)]
````

````python
df00 = df[(df['cgpa_z_score']<3) & (df['cgpa_z_score']>-3)]
df00
````

__Capping__

````python
upper = df['cgpa'].mean()+3*df['cgpa'].std()
lower = df['cgpa'].mean()-3*df['cgpa'].std()
````

````python
df['cgpa'] = np.where(
    df['cgpa']>upper,
    upper,
    np.where(
        df['cgpa']<lower,
        lower,
        df['cgpa']
    )

)


# np.where(
#     condition,
#     (agr shi hui to ky krna ),
#     np.where(
#         condition,
#         (ye wali true hui to ky krna,),
#         last me agr dono nhi hui to
#     )
# )
````
