## __üìå What is Encoding?__

Encoding means converting data from one form into another form so that a computer or machine learning model can understand and process it.

In machine learning, encoding usually means converting text data into numbers.

__üìå Nominal vs Ordinal Data__
‚úÖ 1. Nominal Data

Nominal data contains categories with no order.

‚úî The categories are just names/labels.
‚úî They cannot be ranked.

Examples:

Gender ‚Üí Male, Female

Colors ‚Üí Red, Blue, Green

Cities ‚Üí Delhi, Mumbai, Kolkata

Pets ‚Üí Dog, Cat, Bird

üëâ You cannot say ‚ÄúRed > Blue‚Äù ‚Äî no ordering.

‚úÖ 2. Ordinal Data

Ordinal data contains categories with a meaningful order.

‚úî Order matters
‚úî Difference between ranks is not exact (not numerical)

Examples:

Size ‚Üí Small < Medium < Large

Education ‚Üí School < Graduate < Postgraduate

Satisfaction ‚Üí Poor < Average < Good < Excellent

Ratings ‚Üí 1 ‚≠ê < 2 ‚≠ê < 3 ‚≠ê < 4 ‚≠ê < 5 ‚≠ê

üëâ Order matters, but difference between categories is not equal.

## __1Ô∏è‚É£ Label Encoding__

Definition:
Label Encoding is the process of converting categorical text data into numeric labels. Each unique category is assigned a unique integer.

‚úÖ Why use Label Encoding?

Machine learning algorithms cannot handle text directly.

Label Encoding converts categories into numbers so models can process them.

Useful for ordinal data (where order matters) or small categorical datasets.

Definition:
Ordinal Encoding is a technique to convert categorical data with a meaningful order (ordinal data) into numbers, while preserving the order.

‚úÖ Why use Ordinal Encoding?

ML models cannot work with text labels.

Ordinal data has a rank or order, so we can encode them as integers.

Unlike Label Encoding for nominal data, the numeric values represent order.

```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv(r"C:\Users\Lenovo\Downloads\StudentsPerformance.csv")
df = df.set_index('gender')
df.sample(30)

#  in bakiyo drop islie krna pda taki isse jo colm chahiye vo last me aa jaye 
df = df.drop('writing score',axis=1)
df = df.drop('reading score',axis=1)
df = df.drop('math score',axis=1)



#  testpreparation course lunch are ordinal
# Gender is nominal 
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
x_train
y_train
```

__Tranform to numbers__

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[['group A','group B','group C','group D','group E'],['high school','some high school','some college',"bachelor's degree","associate's degree","master's degree"]])
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
x_train
```
