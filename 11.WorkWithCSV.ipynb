# üìå TSV File (Tab-Separated Values)

## üîπ Definition
TSV (Tab-Separated Values) is a simple text file format used to store data in a table structure, 
where each column is separated by a TAB character.

## üîπ File Extension
.tsv

## üîπ Structure
- Each row represents one record.
- Columns are separated using a TAB space (not comma).
- Data is stored in plain text format.

## üîπ Example

Name	Age	Course	Marks
Piyush	20	B.Tech	85
Rahul	21	B.Sc	78
Anita	19	BCA	92

## üîπ Advantages
- Easy to read and edit.
- Works well with Excel, Google Sheets, and Pandas.
- Useful when data contains commas (because it avoids comma conflicts).

## üîπ Difference Between CSV and TSV

| Feature | CSV | TSV |
|----------|------|------|
| Separator | Comma (,) | Tab |
| File Extension | .csv | .tsv |
| Best Used When | Simple data | Data contains commas |

## üîπ Use in Machine Learning
TSV files are commonly used to store datasets 
that can be loaded into Python using libraries like Pandas.

# ‚öô Important Parameters of read_csv()

---

## 1Ô∏è‚É£ sep Parameter

Used when separator is NOT comma.

Example: Tab Separated File (TSV)

```python
df = pd.read_csv("file.tsv", sep="\t")
```

---

## 2Ô∏è‚É£ header Parameter

If first row is not column name:

```python
df = pd.read_csv("file.csv", header=0)
```

If no header:

```python
df = pd.read_csv("file.csv", header=None)
```

---

## 3Ô∏è‚É£ names Parameter

If file has no column names:

```python
df = pd.read_csv("file.csv", names=["col1","col2","col3"])
```

---

## 4Ô∏è‚É£ index_col Parameter

Set a column as index:

```python
df = pd.read_csv("file.csv", index_col="id")
```

---

## 5Ô∏è‚É£ usecols Parameter

Load only required columns:

```python
df = pd.read_csv("file.csv", usecols=["Name","Age"])
```

---

## 6Ô∏è‚É£ squeeze Parameter (Older versions)

If only one column needed:

```python
df = pd.read_csv("file.csv", usecols=["Name"]).squeeze()
```

---

## 7Ô∏è‚É£ skiprows Parameter

Skip specific rows:

```python
df = pd.read_csv("file.csv", skiprows=[0,2])
```

Skip first N rows:

```python
df = pd.read_csv("file.csv", skiprows=5)
```

---

## 8Ô∏è‚É£ nrows Parameter

Load limited number of rows:

```python
df = pd.read_csv("file.csv", nrows=100)
```

Useful for large datasets.

---

## 9Ô∏è‚É£ encoding Parameter

Fix encoding errors:

```python
df = pd.read_csv("file.csv", encoding="latin1")
```

Default encoding is:
```
utf-8
```

If strange symbols appear ‚Üí change encoding.

---

## üîü on_bad_lines Parameter

Skip corrupted rows:

```python
df = pd.read_csv("file.csv", on_bad_lines="skip")
```

Useful when:
- Some rows have extra columns
- CSV format is broken

---

## 1Ô∏è‚É£1Ô∏è‚É£ dtype Parameter

Force column data type:

```python
df = pd.read_csv("file.csv", dtype={"target": int})
```

Helps:
- Reduce memory usage
- Avoid wrong type detection

---

## 1Ô∏è‚É£2Ô∏è‚É£ parse_dates Parameter

Convert string to datetime:

```python
df = pd.read_csv("file.csv", parse_dates=["date"])
```

Now date column becomes datetime type.

---

### Combining Multiple Columns into Date

```python
df = pd.read_csv("file.csv", parse_dates=[["year","month","day"]])
```

---

## 1Ô∏è‚É£3Ô∏è‚É£ converters Parameter

Apply function to a column while loading:

```python
def short_name(x):
    if x == "Royal Challengers Bangalore":
        return "RCB"
    return x

df = pd.read_csv("file.csv", converters={"team": short_name})
```

---

## 1Ô∏è‚É£4Ô∏è‚É£ na_values Parameter

Define custom missing values:

```python
df = pd.read_csv("file.csv", na_values=["NA","Missing","-"])
```

These values will be treated as NaN.

---

## 1Ô∏è‚É£5Ô∏è‚É£ chunksize Parameter (Very Important for Big Data)

Used when dataset is too large.

```python
df = pd.read_csv("file.csv", chunksize=5000)

for chunk in df:
    print(chunk.shape)
```

Benefits:
- Reduces memory usage
- Handles large datasets
- Process data in parts

# üìä Common Problems in CSV Handling

| Problem | Solution |
|----------|----------|
| Wrong separator | Use `sep` |
| No column names | Use `names` |
| Encoding error | Use `encoding` |
| Extra columns in rows | Use `on_bad_lines` |
| Large file | Use `chunksize` |
| Wrong datatype | Use `dtype` |
| Date not working | Use `parse_dates` |
