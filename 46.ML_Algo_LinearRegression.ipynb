### What is meant by Regression?

Regression is a type of supervised learning used to predict a continuous numerical value based on one or more input variables.

---

### Simple definition
Regression means finding the relationship between variables to predict a real-valued outcome.

---

### Example

| Input (X)        | Output (y)          |
|------------------|---------------------|
| House size       | House price         |
| Study hours      | Exam marks          |
| Temperature      | Electricity usage   |

---

### Mathematical view

y = f(X) + error

Where:
- X = input features
- y = continuous target variable

---

## Types of Linear Regression

### 1️⃣ Simple Linear Regression
- Uses **one independent variable**
- Models a straight-line relationship

Formula:
y = mx + c

Examples:
- Predicting salary from years of experience
- Predicting house price from house size
- Predicting marks from study hours

---

### 2️⃣ Multiple Linear Regression
- Uses **multiple independent variables**
- Models a linear plane or hyperplane

Formula:
y = w₁x₁ + w₂x₂ + ... + wₙxₙ + b

Examples:
- Predicting house price using size, location, and number of rooms
- Predicting car price using mileage, engine power, and age

---

### 3️⃣ Polynomial Linear Regression
- Models **non-linear relationships**
- Still linear in parameters

Formula:
y = w₀ + w₁x + w₂x² + ... + wₙxⁿ

Examples:
- Modeling growth rate vs time
- Predicting stock trends with curved patterns
- Modeling temperature vs energy consumption

---

### Why Polynomial Regression is still Linear?
- Linear with respect to weights
- Non-linearity comes from feature transformation

---

### Regression vs Classification

| Regression            | Classification        |
|-----------------------|-----------------------|
| Predicts numbers      | Predicts categories   |
| Continuous output     | Discrete output       |
| Price, temperature   | Spam / Not spam       |

---

### One-line answer (Interview-ready)

Regression is a supervised learning technique used to predict continuous values by learning relationships between variables.

## Simple Linear Regression (Explained with Example)

### What is Simple Linear Regression?

Simple Linear Regression is a regression technique used to predict a continuous output using one independent variable.

It assumes a linear relationship between input and output.

---

### Mathematical Formula

y = mx + c

Where:
- x = independent variable (input)
- y = dependent variable (output)
- m = slope (change in y when x increases by 1)
- c = intercept (value of y when x = 0)

---

## Example: Predicting Exam Marks from Study Hours

### Given Data

| Study Hours (x) | Marks (y) |
|----------------|----------|
| 2              | 40       |
| 4              | 55       |
| 6              | 65       |
| 8              | 80       |

---

### Step 1: Understand the Relationship
- As study hours increase, marks increase
- Relationship appears linear

---

### Step 2: Fit a Line

Assume the learned equation is:

y = 5x + 30

This means:
- For every additional hour studied, marks increase by 5
- If a student studies 0 hours, expected marks = 30

---

### Step 3: Make a Prediction

If a student studies 7 hours:

y = 5(7) + 30 = 65

Predicted marks = 65

---

## Geometric Interpretation

- Data points are plotted on a graph
- Regression finds the best-fit straight line
- Errors are vertical distances from points to the line

---

## Key Points to Remember

- Uses one feature
- Output is continuous
- Minimizes sum of squared errors
- Simple but powerful

---

## One-line Interview Answer

Simple linear regression predicts a continuous output using one input variable by fitting a best-fit straight line.

## Best-Fit Line in Simple Linear Regression

### What is the Best-Fit Line?

In simple linear regression, the **best-fit line** is the straight line that represents the relationship between the input variable (x) and the output variable (y) in the most accurate way.

---

### Mathematical Form

y = mx + c

Where:
- m = slope of the line
- c = intercept

---

### How is the Best-Fit Line Determined?

The best-fit line is the line that **minimizes the sum of squared errors (SSE)**.

Objective:

min Σ (yᵢ − ŷᵢ)²

This method is called **Ordinary Least Squares (OLS)**.

---

### Intuition

- Each data point has a vertical distance from the line
- The best-fit line minimizes the total squared vertical distances
- Squaring prevents positive and negative errors from canceling out

---

### One-Line Interview Answer

The best-fit line is the line that minimizes the sum of squared differences between actual and predicted values in simple linear regression.
