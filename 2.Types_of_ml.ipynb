__1. Supervised Learning__

. Learns from labeled data (input + correct output).

. Examples: classification, regression.

__2. Unsupervised Learning__

The model learns from unlabeled data and finds hidden patterns.

Use cases: clustering, grouping, pattern discovery.

Examples: customer segmentation, grouping similar images.

__3. Semi-Supervised Learning__

A mix of labeled + unlabeled data.

Used when labeling all data is expensive.

Examples: medical images, fraud detection.

__4. Reinforcement Learning__

The model learns by trial and error, receives rewards for correct actions.

Examples: robotics, self-driving cars, game-playing AI.

__A. Supervised Learning Types__

1.Classification – Predict categories

2.Regression – Predict continuous values

__B. Unsupervised Learning Types__

1.Clustering – Group similar data

2.Association Rule Learning – Find relations between items

3.Dimensionality Reduction – Reduce features (PCA, LDA)

4.Anomaly Detection – Find unusual data

__C. Reinforcement Learning Types__

1.Positive Reinforcement

2.Negative Reinforcement

3.Q-Learning

4.Policy Gradient Methods
